```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hr=pd.read_csv(r"C:\ksr\data science\DS_batch1\datasets\HR_comma_sep.csv")
```

```python
hr=hr.rename(columns=lambda x:x.strip())
```

```python
hr1=pd.get_dummies(hr)
```

```python
x_train=hr1.drop(["left"],axis=1)
y_train=hr1["left"]
```

```python
from sklearn import model_selection

kfold=model_selection.KFold(n_splits=3)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
model=BaggingClassifier(warm_start=True)
model.fit(x_train,y_train)
```

```python
help(BaggingClassifier)
```

```python
results=model_selection.cross_val_score(model,x_train,y_train,cv=kfold)
```

```python
results.mean()
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=20,n_informative=15,n_redundant=5,random_state=5)
```

```python
X
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
model=BaggingClassifier()
cv=RepeatedStratifiedKFold(n_splits=10,n_repeats=3,random_state=1)
n_scores=model_selection.cross_val_score(model,X,y,scoring="accuracy",cv=cv)
print("accuracy: %.3f(%.3f)" % (np.mean(n_scores),np.std(n_scores)))
```
