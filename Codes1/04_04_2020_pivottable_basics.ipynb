```python
rng=np.random.RandomState(0)
df=pd.DataFrame({"key":["A","B","C","A","B","C"],"data1":range(6),"data2":rng.randint(0,10,6)},columns=["key","data1","data2"])
```

```python
df
```

```python
df.groupby("key").aggregate(["min",np.median,max,sum])
```

```python
def filter_func(x):
    return x["data2"].std()>4
```

```python
print(df.groupby("key").filter(filter_func))
```

```python
df
```

```python
df.groupby("key").transform(lambda x:x-x.mean())
```

```python
def normy_by_data2(x):
    x["data1"]/=x["data2"].sum()
    return x
```

```python
print(df.groupby("key").apply(normy_by_data2))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.shape
```

```python
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].count()
```

```python
titanic.groupby(["sex","class"])[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
titanic.groupby(["sex","class"])[["survived"]].mean().unstack()
```

```python
titanic.pivot_table("survived",index="sex",columns="class")
```

```python
pd.pivot_table(index="sex",columns="class",aggfunc={"survived":np.mean},data=titanic)
```

```python
titanic.pivot_table(index="sex",columns="class",aggfunc={"survived":len})
```

```python
titanic.head()
```

```python
age=pd.cut(titanic["age"],[0,18,80])
titanic.pivot_table("survived",["sex",age],"class")
```

```python
titanic.pivot_table(index=["sex",age],columns="class",aggfunc={"survived":np.mean})
```

```python
age=pd.qcut(titanic["age"],q=[0,0.25,0.5,0.75,1])
titanic.pivot_table(index=["sex",age],columns="class",aggfunc={"survived":np.mean})
```

```python
age=pd.qcut(titanic["age"],q=4)
titanic.pivot_table(index=["sex",age],columns="class",aggfunc={"survived":len})
```

```python
titanic["age"].describe()
```

```python
titanic.head()
```

```python
titanic.pivot_table("survived",index="sex",columns="class",margins=True)
```

```python
titanic.pivot_table("survived",index="pclass",columns="sex",aggfunc="sum")
```

```python
import matplotlib.pyplot as plt
sns.set()
titanic.pivot_table("survived",index="pclass",columns="sex",aggfunc="sum").plot()
plt.ylabel("total counts")
```
