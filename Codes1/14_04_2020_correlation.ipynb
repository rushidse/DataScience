```python
def corr_cal(data,x,y):
    x1=sum((data[x]-data[x].mean())*(data[y]-data[y].mean()))
    length=(len(data[x])-1)
    cov=x1/length
    std1=data[x].std()
    std2=data[y].std()
    stdr=std1*std2
    corr= cov/stdr
    return corr
```

```python
corr_cal(data=air1,x="Promotion_Budget",y="Passengers")
```
